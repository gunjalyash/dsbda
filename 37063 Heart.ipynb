```python
# import pandas library
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.linear_model import LogisticRegression
```

```python
# Reading csv file
data = pd.read_csv("Heart.csv")
# creating a dataframe of it
data = pd.DataFrame(data)
data.head()
```

```python
# Datatype of each column
data.dtypes
```

```python
# Information about each column data
data.info()
```

```python
# Count ,min,max ,etc of each column
data.describe()
```

```python
#Drop unnnamed column
data.drop("Unnamed: 0",axis=1,inplace=True)
data.head()
```

```python
#Finding null values in each column
data.isna().sum()
```

```python
ca_mean = data['Ca'].mean()
data['Ca'] = data['Ca'].fillna(ca_mean)
data.isna().sum()
```

```python
ca_mode = data['Thal'].mode()[0]
data['Thal'] = data['Thal'].fillna(ca_mode)
data.isna().sum()
```

```python
#Finding number of zeros in each column
for i in data:
  count = (data[i] == 0).sum()
  print('Zeros in column',i,'->', count)
```

```python
#Label Encoding the 'AHD' column
le1=LabelEncoder()
data['AHD']=le1.fit_transform(data['AHD'])
data.head()
```

```python
le1=LabelEncoder()
data['Thal']=le1.fit_transform(data['Thal'])
data.head()
```

```python
le1=LabelEncoder()
data['ChestPain']=le1.fit_transform(data['ChestPain'])
data.head()
```

```python
# splitting data using train test split
x = data.drop(['AHD'], axis=1)
y = data.AHD
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)

x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
#Classification model using Decision Tree
from sklearn.tree import DecisionTreeClassifier
tc=DecisionTreeClassifier(criterion='entropy')
tc.fit(x_train,y_train)
y_train_pred=tc.predict(x_train)

print("Training Accuracy Score :",accuracy_score(y_train_pred,y_train))
print("Training Confusion Matrix  :",confusion_matrix(y_train_pred,y_train))
```

```python
y_test_pred=tc.predict(x_test)

print("Accuracy Score :",accuracy_score(y_test_pred,y_test))
print("Confusion Matrix  :",confusion_matrix(y_test_pred,y_test))
```

```python
# Classification model using Logistic Regression
lg = LogisticRegression(random_state=0)
lg.fit(x_train, y_train)
y_train_pred = lg.predict(x_train)
print("Training Accuracy Score :",accuracy_score(y_train_pred,y_train))
print("Training Confusion Matrix  :",confusion_matrix(y_train_pred,y_train))
```

```python
y_test_pred = lg.predict(x_test)
print("Testing Accuracy Score :",accuracy_score(y_test_pred,y_test))
print("Testing Confusion Matrix  :",confusion_matrix(y_test_pred,y_test))
```
